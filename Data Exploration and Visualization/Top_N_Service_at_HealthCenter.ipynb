# Q2: What are the Top-N reportable cases or services in the healthcenter?

```python
topndf = pd.read_csv("FinalSelectedHCReportTotal-Table 1.csv", encoding='utf-8')
```

```python
topndf.head(2)
```

```python
#count by 
topndf.groupby(['Service_Name','Satus']).size().unstack().plot(kind='barh',stacked=True,figsize=(10,20))
plt.legend(bbox_to_anchor=(1, 1), loc='best')
plt.xlabel("No. of service delivery by category", fontsize=20)
plt.ylabel("Service",fontsize=14)
plt.tick_params(axis='both',labelsize=14)
plt.show()
```

```python
#count by 
topndf.groupby(['Service_Name','Yearly_Total']).size().unstack().plot(kind='barh',stacked=True,edgecolor='k',figsize=(20,30))
plt.legend(bbox_to_anchor=(1, 1),prop={'size': 30})
plt.xlabel("No. of service delivery by category", fontsize=20)
plt.ylabel("Service",fontsize=30)
plt.tick_params(axis='both',labelsize=30)
plt.show()
```
